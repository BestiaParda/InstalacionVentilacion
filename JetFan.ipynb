# Instalación de Ventiladores de chorro (JetFan)


## Objeto
Se trata de establecer los criterios de instalación de ventiladores tipo JetFan.

## Hipótesis de Partida
Se trata de una edificación convencional:
    
- El fluido será aire a temperatura de $20ºC$
- La instalación se realiza en garajes

## Empuje en descarga libre
El empuje producido en un fluido viene determinado por:
     $$T=\rho V Q $$
    
- $\rho$ es la densidad del fluido ($kg/m^3$)
- V velocidad del fluido en el punto de descarga ($m/s$)
- Q caudal del fluido en el punto de descarga ($m^3/s$)

Si sustituimos el caudal en función del area de descarga tenemos:
         $$T=\rho V^2 \pi r^2 $$        
         $$T=\frac{\rho V^2 \pi \phi^2}{4} $$

- r el radio del ventilador en el punto de descarga ($m$)
- $\phi$ el diametro del ventilador en el punto de descarga ($m$)

Fijado el diametro del ventilador necesitamos el siguiente empuje en el ventilador para que el fluido alcance las siguientes velocidades:
    


```python
densidad=1.2
T(v,d)=densidad*v^2*pi*(d/1000)^2/4
p1=plot(T(x,300), (x, 1, 30),color='red',axes_labels =['$V(m/s)$','$T(N)$'],figsize=8,legend_label='$\phi=300$')
p2=plot(T(x,350), (x, 1, 30),color='navy',legend_label='$\phi=350$')
p3=plot(T(x,400), (x, 1, 30),color='green',legend_label='$\phi=400$')
show(p1+p2+p3)
```

## Equivalencia de Diámetros

Para establecerer el diametro del ventilador equivalente con descarga con sección cuadrada mantenemos el empuje:
$$T=\frac{\rho V_1^2 \pi \phi^2}{4}=\rho V_2^2l^2 $$

- $V_1$ velocidad del fluido en el ventilador de seccion circular ($m/s$)
- $V_2$ velocidad del fluido en el ventilador de seccion cuadrada ($m/s$)
- $l$ es el lado del area del ventilador de sección cuadrada ($m$)

Si fijamos el lado del ventilador de sección cuadrada, la velocidad que debemos fijar será:
$$V_2=\sqrt{\frac{V_1^2\pi \phi^2}{4l^2}}$$
